# 01 Setup (CPU)

Cel: sprawdzic srodowisko i przygotowac stabilny start projektu.


```python
# Krok 1: importujemy biblioteki potrzebne do startu
import os
import random
import numpy as np
import torch

# Krok 2: ustawiamy seed, aby wyniki byly powtarzalne
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)

# Krok 3: wymuszamy CPU i ustawiamy liczbe watkow
torch.set_num_threads(max(1, os.cpu_count() // 2))
device = torch.device("cpu")

# Krok 4: szybki sanity check wersji i urzadzenia
print("Torch version:", torch.__version__)
print("Device:", device)
print("Threads:", torch.get_num_threads())
```
