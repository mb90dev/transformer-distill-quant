```python
# Krok 1: importy
import json
import os
import pandas as pd
```

```python
# Krok 2: helper do bezpiecznego czytania json
def load_json_if_exists(path):
    if os.path.exists(path):
        with open(path, "r", encoding="utf-8") as f:
            return json.load(f)
    return None

baseline = load_json_if_exists("../outputs/metrics_student_baseline.json")
distilled = load_json_if_exists("../outputs/metrics_student_distilled.json")
bench = load_json_if_exists("../outputs/bench_results.json")
```

```python
# Krok 3: skladamy prosta tabele wynikow
rows = []

if baseline is not None:
    rows.append({
        "model": "student_fp32_baseline",
        "test_accuracy": baseline["test"]["accuracy"],
        "test_f1_macro": baseline["test"]["f1_macro"],
    })

if distilled is not None:
    rows.append({
        "model": "student_fp32_distilled",
        "test_accuracy": distilled["test"]["accuracy"],
        "test_f1_macro": distilled["test"]["f1_macro"],
    })

metrics_df = pd.DataFrame(rows)
display(metrics_df)
```

```python
# Krok 4: podglad benchmarku
if bench is not None:
    bench_df = pd.DataFrame(bench).T.reset_index().rename(columns={"index": "variant"})
    display(bench_df)
else:
    print("Brak bench_results.json - uruchom notebook 07")
```

```python
# Krok 5: gotowy bullet do CV
cv_bullet = (
    "Implemented a custom PyTorch training loop for transformer distillation "
    "(CrossEntropy + KL with temperature), applied INT8 dynamic quantization "
    "for CPU inference, and benchmarked p50/p95 latency and throughput across "
    "Teacher FP32, Student FP32, and Student INT8 variants."
)
print(cv_bullet)

with open("../outputs/cv_bullet.txt", "w", encoding="utf-8") as f:
    f.write(cv_bullet + "\\n")
```
